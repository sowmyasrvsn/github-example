```python
import pandas as pd    
page = 'https://en.wikipedia.org/wiki/List_of_postal_codes_of_Canada:_M'    
tables = pd.read_html(page)     
tables[0].head(20) 
```

```python
tables[0] = tables[0][tables[0].Borough != 'Not assigned']
tables[0].head(20)
```

```python

# Create a list of indices of the cells where the value of Column2 is "Not assigned"
idx_to_change = tables[0].loc[tables[0]['Neighbourhood'] == "Not assigned"].index

# Iterate the list of indicies and set Column2 to the value of Column1 at the given index
for i in idx_to_change:
    tables[0].iloc[i, 1] = tables[0].postcodes.iloc[i, 0] 
    
tables[0]
```

```python
df = tables[0].groupby(['Postcode','Borough'], sort=False).agg( ','.join)
print(df)
```

```python
df1 = pd.DataFrame(df) 
  
print(df1)  
```
