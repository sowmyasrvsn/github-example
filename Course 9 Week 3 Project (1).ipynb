```python
import pandas as pd    
page = 'https://en.wikipedia.org/wiki/List_of_postal_codes_of_Canada:_M'    
tables = pd.read_html(page)     
tables[0].head(20) 
```

```python
tables[0] = tables[0][tables[0].Borough != 'Not assigned']
tables[0].head(20)
```

```python

# Create a list of indices of the cells where the value of Column2 is "Not assigned"
idx_to_change = tables[0].loc[tables[0]['Neighbourhood'] == "Not assigned"].index

# Iterate the list of indicies and set Column2 to the value of Column1 at the given index
for i in idx_to_change:
    tables[0].iloc[i, 1] = tables[0].postcodes.iloc[i, 0] 
    
tables[0]
```

```python
df = tables[0].groupby(['Postcode','Borough'], sort=False).agg( ','.join).reset_index()
df.head()
```

```python
df1 = pd.DataFrame(df) 
df1.reset_index()
print(df1)  
```

```python
df2=pd.read_csv('http://cocl.us/Geospatial_data')
df2.rename(columns={"Postal Code":"Postcode"},inplace=True)
df2.head()
```

```python
df3=df.merge( df2, on='Postcode', how='inner')
df3.head(5)
```

```python
dfToronto=df3[df3['Borough'].str.contains("Toronto")]
dfToronto.head()

```

```python

map_clusters = folium.Map(location=[43.676357, -79.293031], zoom_start=10)


for index, row in dfToronto.iterrows(): 

    folium.Marker(

    [row['Latitude'], row['Longitude']],

    radius=10,

    color='red',

    popup="M4N",

    fill = True,

    fill_color = 'blue'
    

map_clusters
```
